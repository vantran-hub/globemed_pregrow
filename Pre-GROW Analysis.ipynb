```python
#Convert strings to numeric values

preg = preg.replace({"Never": 1, "Sometimes": 3, "Frequently": 4, "Always": 5})
```

```python
#How many chapters did participate in the pre-GROW training survey?

n = len(pd.unique(preg["GlobeMed Chapter:"]))
  
print("No.of chapters :", 
      n)

preg["GlobeMed Chapter:"].value_counts()
```

```python
#Bar plot of chapters' participation in the Pre-GROW survey

preg["GlobeMed Chapter:"].value_counts()[:10].sort_values().plot(kind = 'barh', color='red')
```

```python
#Create a dataset for GROW coordinators

preg_coor = preg[preg["GlobeMed Role:"] == "GROW Coordinator"]
print(preg_coor.shape)
preg_coor.describe()
```

```python
#Create a dataset for GROW interns

preg_int = preg[preg["GlobeMed Role:"] == "GROW Intern"]
print(preg_int.shape)
preg_int.describe()
```

```python
#Which skills are interns most and least confident about before the GROW training?

preg_int1 = preg_int.drop(columns=['First Name:','Last Name:','GlobeMed Role:','GlobeMed Chapter:'])
mp1 = preg_int1.mean()
mp1.sort_values()
```

```python
#Export dataset for visualization

mp1.to_csv('globemed_mp1.csv')
```

# Conclusion

1. 87 responses were recorded. Among them 16 were from GlobeMed coordinators and 71 were from GlobeMed interns.
2. There were 24 out of 28 chapters participating in the pre-GROW survey. The most active chapter in the survey is GlobeMed at Northwestern University with 9 individual participants.
3. Data suggests a strong baseline on the ability of students to practice skills related to each GROW goal before the GROW experience. Skills were rated on the scale of 1 to 5. 
4. Most students reported strong confidence on general leadership skills such as having an open perspective, identify & grow strengths, and mutual accountability. 
5. The skills with lowest points reported were strong friendships with students' chapter's partner contacts and community members, as well as skills related to GROW Goal #3: Partnership Reflection & Planning, including knowledge about strengths and areas of growth for students' partnership from this past year and clear picture of partnerships’ goals for the upcoming year.
